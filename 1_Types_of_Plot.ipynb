# Interactive Plots

Normally, there are 2 types of plot that can be done on IPython notebooks using Matplotlib:

`%matplotlib notebook` will lead to interactive plots embedded within the notebook\
`%matplotlib inline` will lead to static images of your plot embedded in the notebook

However, for Colab, we cannot use the `%matplotlib notebook` command. Rather, we need to use the mpld3 library

More can be referenced from: https://colab.research.google.com/drive/1tXS-kwZUWgkJqYXKxglg67yGQmenSl8d#scrollTo=RqPMGGp7gZiC


```
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
%matplotlib inline

x = [1,2,3]
y = [4,5,6]

plt.plot(x,y) 
mpld3.display()
```

# Legends, Title and Labels

```
x1 = [4,6,9]
y1 = [100,200,600]

x2 = [50,60,110]
y2 = [25,30,90]

plt.plot(x1,y1, label ='First line') 
plt.plot(x2,y2, label ='Second line') 

plt.xlabel("Plot Number")
plt.ylabel("Important Variable")
plt.title("Interesting Graph\nCheck it out") #For multiple line title
plt.legend()
plt.show()
```

# Bar Charts and Histograms

```
import numpy as np

x1 = np.random.randint(0,10, size=5)
y1 = np.random.randint(100, size=5)

x2 = np.random.randint(10, 20, size=5)
y2 = np.random.randint(100, size=5)

plt.bar(x1,y1, label ='Bar 1', color='r')
plt.bar(x2,y2, label ='Bar 2', color='c')

plt.xlabel("Plot Number")
plt.ylabel("Random Numbers")
plt.title("Random interesting Graph\nCheck it out") #For multiple line title
plt.legend()
plt.show()

```

```
population_ages = np.random.randint(130, size=20)
population_ages_normal = np.random.normal(45, 5, 20) #Outputs gaussian distribution: Mean,SD, no. of samples
bins = np.linspace(0,130,14)

plt.hist(population_ages,bins, histtype='bar', rwidth=0.5)

plt.xlabel("Plot Number")
plt.ylabel("Random Numbers")
plt.title("Random interesting Graph\nCheck it out") #For multiple line title
plt.legend()
plt.show()
```

# Scatter Plots

The marker size, color and shape can be altered. More on:
https://matplotlib.org/api/markers_api.html?highlight=marker

```
x = np.random.randint(130, size=10)
y = np.random.randint(50, size=10)

plt.scatter(x,y, label='scatter', color = 'g', marker = 'd', s=100)

plt.xlabel("Plot Number")
plt.ylabel("Random Numbers")
plt.title("Random interesting Graph\nCheck it out") #For multiple line title
plt.legend()
plt.show()
```

# Stack Plots

```
day = [3,4,5,6]

sleeping = [7,8,9,11]
eating = [3,2,2,3]
working = [8,9,10,4]
learning = [6,5,3,6]

plt.stackplot(day, sleeping, eating, working, learning, labels=('sleeping', 'eating', 'working','learning'))

plt.xlabel("Plot Number")
plt.ylabel("Random Numbers")
plt.title("Random interesting Graph\nCheck it out") #For multiple line title
plt.legend()
plt.show()
```

# Pie Chart


```
day = [3,4,5,6]

sleeping = [7,8,9,11]
eating = [3,2,2,3]
working = [8,9,10,4]
learning = [6,5,3,6]

slices = [6,7,4,8]
plt.pie(slices, labels=('sleeping', 'eating', 'working','learning'), 
        explode=[0,0,0.2,0], autopct='%2.1f%%',
        startangle=180, radius=1.5,
        labeldistance = 1.1, shadow=True)

plt.xlabel("Plot Number")
plt.ylabel("Random Numbers")
plt.title("Random interesting Graph\nCheck it out") #For multiple line title
plt.legend()
plt.show()
```
